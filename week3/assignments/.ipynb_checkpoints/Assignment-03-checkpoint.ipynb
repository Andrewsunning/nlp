```python
from scipy.spatial.distance import cosine
```

```python
X = dataset.data[:,4:6]
y = dataset.target
```

```python
def model(X,y):
    """
    建立一个KNN模型
    --------
    Return：
        list,返回由每条记录的X和y组成的元组列表
    --------
    Example：
        dataset = load_boston()
        X = dataset.data[:,4:6]
        y = dataset.target
        model(X,y)[:2]
    """
    return [(X_i,y_i) for X_i,y_i in zip(X,y)]
model(X,y)[:2]
```

```python
def distance(X_1,X_2):
    """
    计算向量X_1和X_2z的余弦相似度
    --------
    Return：
        int
    --------
    Example：
        distance([1, 1], [2, 2])
    """
    return cosine(X_1,X_2)
distance([1, 1], [2, 2])
```

```python
import numpy as np
```

```python
def predict(x, k=5):
    """
    预测x的预测值（回归任务取平均）
    --------
    Return：
        int，向量x的预测值
    --------
    Example：
        predict([0.499, 5.841])
    """
    most_similar = sorted(model(X,y), key=lambda x_i:distance(x_i[0][0],x))[:k]    # 注意：most_simalar是一个列表元祖，e.g.[(array([0.469, 6.421]), 21.6)]
    most_targets = [item[1] for item in most_similar]
    return np.average(most_targets)
print(predict([0.499, 5.841]))
```

```python
from collections import Counter
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
def entropy(elements):
    """
    计算序列elements的信息熵
    -------
    Return:
        int,序列elements的信息熵
    -------
    Example:
        elements = [1,1,0,0]
        entropy(elements)
    """
    counter = Counter(elements)    # 计数器
    n = len(elements)
    probs = []
    for item in set(elements):
        probs.append(counter[item]/n)
    return -sum([p*np.log2(p) for p in probs])
elements = [1,1,0,0]
entropy(elements)
```

```python
mock_data = {
    'gender':['F', 'F', 'F', 'F', 'M', 'M', 'M'],
    'income': ['+10', '-10', '+10', '+10', '+10', '+10', '-10'],
    'family_number': [1, 1, 2, 1, 1, 1, 2],
    'bought': [1, 1, 1, 0, 0, 0, 1],
}

dataset = pd.DataFrame(mock_data)
dataset
```

```python
# @pysnooper.snoop()
def findMinEntropty_Spliter(train_data:pd.DataFrame, target:str):
    """
    计算数据集train_data每个特征的信息熵，并返回熵最小（信息增益最大）的特征及取值
    -------
    Return:
        tuple,特征及其分裂值
    -------
    Example:
        findMinEntropty_Spliter(dataset, "bought")
    """
    featuers = set(train_data.columns) - set([target])
    spliter = None
    min_entropy = float("inf")
    for feature in featuers:
        values = set(train_data[feature])    # 某一feature能取到的不同值的集合
        for v in values:
            # here
            sub_spliter_1 = train_data.loc[train_data[feature]==v][target]
            entropy_1 = entropy(sub_spliter_1)
            
            sub_spliter_2 = train_data.loc[train_data[feature]!=v][target]
            entropy_2 = entropy(sub_spliter_2)
            
            entropy_feature = entropy_1+entropy_2
            if entropy_feature<min_entropy:
                min_entropy = entropy_feature
                spliter = feature,v
    return spliter

findMinEntropty_Spliter(dataset, "bought")
```

```python
sub_dataset_by_family_number_1 = dataset.loc[dataset["family_number"]==1]
sub_dataset_by_family_number_2 = dataset.loc[dataset["family_number"]!=1]
print(findMinEntropty_Spliter(sub_dataset_by_family_number_1, "bought"))
```

1. In our course and previous practice, we complete some importance components of Decision Tree. In this problem, you need to build a **completed** Decision Tree Model. You show finish a `predicate()` function, which accepts three parameters **<gender, income, family_number>**, and outputs the predicated 'bought': 1 or 0.  (20 points)

```python
"""
Decision Tree结构:
    findMinEntropty_Spliter(train_data:pd.DataFrame, target:str)
    splitDataSet(dataset:pd.DataFrame, spliter:str, value)
    createTree(dataset, features:list, target:str)
    predict(myTree:dict, features:list, X:list)
"""
```

```python
# @pysnooper.snoop()
def findMinEntropty_Spliter(train_data:pd.DataFrame, target:str):
    """
    计算数据集train_data每个特征的信息熵，并返回熵最小（信息增益最大）的特征及取值
    -------
    Return:
        tuple,特征及其分裂值
    -------
    Example:
        findMinEntropty_Spliter(dataset, "bought")
    """
    featuers = set(train_data.columns) - set([target])
    spliter = None
    min_entropy = float("inf")
    for feature in featuers:
        values = set(train_data[feature])    # 某一feature能取到的不同值的集合
        for v in values:
            # here
            sub_spliter_1 = train_data.loc[train_data[feature]==v][target]
            entropy_1 = entropy(sub_spliter_1)
            
            sub_spliter_2 = train_data.loc[train_data[feature]!=v][target]
            entropy_2 = entropy(sub_spliter_2)
            
            entropy_feature = entropy_1+entropy_2
            if entropy_feature<min_entropy:
                min_entropy = entropy_feature
                spliter = feature
    return spliter

findMinEntropty_Spliter(dataset, "bought")
```

```python
import pysnooper
```

```python
from collections import Counter

def splitDataSet(dataset:pd.DataFrame, splitFeature:str, value):
    """
    将数据集dataset按照特征spliter=value切分出数据子集
    ------
    Return:
        pd.DataFrame,表示特征spliter=value的特征子集
    ------
    Examples：
        splitDataSet(dataset,"income","+10")
    """
    retDataset = dataset.loc[dataset[splitFeature]==value]
    retDataset.drop(columns=splitFeature, inplace=True)    # 删除数据集中用户划分的特征列
    return retDataset

# @pysnooper.snoop()
def createTree(dataset, features:list, target:str):    # 注意features包含target特征
    """
    创建一颗决策树，以字典数据结构来保存
    ------
    Return:
        dict,表示决策树的划分特征及其取值
    ------
    Examples：
        createTree(dataset.loc[dataset["bought"]==0],features=dataset.columns, target="bought")
    """
    classList = dataset[target].tolist()
    if classList.count(classList[0])==len(classList):    # 如果数据子集中的样本的类别都为同一类，则停止切割，返回该类别
        return classList[0]
    if len(features)==1:    # 如果特征子集仅剩一个target特征，则采用投票法返回出现次数最多的类别
        return Counter(classList).most_common(1)[0][0]
    
    bestFeature = findMinEntropty_Spliter(dataset, "bought")
    features.remove(bestFeature)    # 删除特征集合中用户划分的特征
    myTree = {bestFeature:{}}
    uniqueValues = set(dataset[bestFeature])
    for val in uniqueValues:
        subFeatures = features[:]    # 复制一个list
        myTree[bestFeature][val] = createTree(splitDataSet(dataset, bestFeature,value=val),subFeatures, target=target)
    return myTree

# splitDataSet(dataset,"income","+10")
# createTree(dataset.loc[dataset["bought"]==0],features=dataset.columns, target="bought")
allFeatures = list(dataset.columns)
myTree = createTree(dataset,features=allFeatures, target="bought")    # 树的key为特征名及其取值，特征值为子树或者类别myTree = 
# {'family_number': {1: {'income': {'+10': {'gender': {'M': 0, 'F': 1}},
#     '-10': 1}},
#   2: 1}}
```

```python
def predict(myTree:dict, features:list, X:list):
    """
    利用创建的决策树对新数据X进行预测
    ------
    Return:
        int,表示X的预测结果
    ------
    Examples：
        predict(myTree, features,["M","-10",1])
    """
    firstFeature = list(myTree.keys())[0]
    secondDict = myTree[firstFeature]
    featureIndex = features.index(firstFeature)
    for key in secondDict.keys():
        if X[featureIndex]==key:
            if isinstance(secondDict[key],dict):    # 出错：这里应该判断secondDict的key对应的value是否为字典类型
                predictLabel = predict(secondDict[key], features, X)
            else:
                predictLabel = secondDict[key]
    return predictLabel

# predict(myTree, features,["M","+10",1])
features = list(dataset.columns)
features.remove("bought")
print("预测结果为:{}".format(predict(myTree, features,["M","-10",1])))
```

2. 将上一节课(第二节课)的线性回归问题中的Loss函数改成"绝对值"，并且改变其偏导的求值方式，观察其结果的变化。(19 point)

```python
# 梯度下降实现--基于Boston数据集
from sklearn.datasets import load_boston
import random
```

```python
dataset = load_boston()
print(dataset.data.shape)
x,y = dataset["data"], dataset["target"]
```

```python
X_rm = x[:,5]   # 取出第五列
plt.scatter(X_rm, y)    # 绘制散点图
plt.show()
```

```python
def predict_price(x, k, b):
    """定义预测函数"""
    return k*x+b

def abs_loss(y, y_hat):
    """定义平方损失函数"""
    n = len(y)
    loss_num = sum(np.abs(y_i-y_hat_i) for y_i,y_hat_i in zip(list(y), list(y_hat)))/n
    return loss_num
abs_loss([1,2,1,1,1,1,2],[1,2,1,1,1,1,-5])
```

```python
def partial_derivative_k(x, y, y_hat):    # 绝对值损失函数对参数k求导
    """定义参数k的偏导数"""
    n = len(y)
    gradient = 0
    for x_i,y_i,y_hat_i in zip(list(x), list(y), list(y_hat)):
        if y_i>y_hat_i:
            gradient+=-x_i
        elif y_i<y_hat_i:
            gradient+=x_i
        else:
            gradient+=0
    return gradient/n

def partial_derivative_b(y, y_hat):    # 绝对值损失函数对参数b求导
    """定义参数b的偏导数"""
    n = len(y)
    gradient = 0
    for y_i,y_hat_i in zip(list(y), list(y_hat)):
        if y_i>y_hat_i:
            gradient+=-1
        elif y_i<y_hat_i:
            gradient+=1
        else:
            gradient+=0
    return gradient/n
# print(partial_derivative_k([1,2], [1,2],[2,0]))
print(partial_derivative_b([1,2],[2,3]))
```

```python
# 初始化参数
k = random.random()*200-100
b = random.random()*200-100

learning_rate = 1e-3
iterations = 20000
sum_loss = []

for i in range(iterations):
    # 计算预测值
    y_hat = [predict_price(x,k,b) for x in X_rm]
    # 计算损失
    loss = abs_loss(y,y_hat)
    sum_loss.append(loss)
    print("Iteration {}, the loss is {}, parameters k is {} and b is {}".format(i,loss,k,b))
    # 更新参数
    k = k-learning_rate*partial_derivative_k(X_rm,y,y_hat)
    b = b-learning_rate*partial_derivative_b(y,y_hat)
break_k = k
break_b = b
```

```python
# 损失变化曲线
plt.plot(sum_loss)
```

```python
# 预测值与真实值
plt.scatter(X_rm,y,color="red")
plt.scatter(X_rm,[predict_price(x,break_k,break_b) for x in X_rm], color="blue")
```
